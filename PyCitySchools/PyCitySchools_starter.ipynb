```python
# format currency $ in dataframe
def format (x):
    return "${:,.2f}".format(x)

def format_1(y):
    return '{:,.2f}%'.format(y)
```

```python
school_data = school_data.sort_values(by=['school_name'])
school_data['budget_average'] = school_data['budget'] / school_data['size']
school_data['budget'] = school_data['budget'].apply(format)
# school_data['budget_average'] = school_data['budget_average'].apply(format)
```

```python
student_data_bySchool = student_data.groupby(['school_name'])

# Calculate Average Maths Score
Total_maths_score = student_data_bySchool['maths_score'].sum()
student_amount = student_data_bySchool['student_name'].count()
average_maths_score = round (Total_maths_score / student_amount,2)
average_maths_score = average_maths_score.to_frame(name='average maths score')
per_school_summary = pd.merge(school_data, average_maths_score,on='school_name')
```

```python
Passing_math_student_df = student_data.loc[school_data_complete_df['maths_score'] >= 50]
Passing_math_student_bySchool = Passing_math_student_df.groupby(['school_name'])
Passing_math_percentage_bySchool = round (Passing_math_student_bySchool['maths_score'].count() / student_amount * 100,2)
Passing_math_percentage_bySchool = Passing_math_percentage_bySchool.to_frame(name='% Pass Maths')
per_school_summary = pd.merge(per_school_summary, Passing_math_percentage_bySchool,on='school_name')
```

```python
# Calculate Average Reading Score
Total_reading_score = student_data_bySchool['reading_score'].sum()
student_amount = student_data_bySchool['student_name'].count()
average_reading_score = round (Total_reading_score / student_amount,2)
average_reading_score = average_reading_score.to_frame(name='average reading score')
per_school_summary = pd.merge(per_school_summary, average_reading_score,on='school_name')
```

```python
Passing_reading_student_df = student_data.loc[school_data_complete_df['reading_score'] >= 50]
Passing_reading_student_bySchool = Passing_reading_student_df.groupby(['school_name'])
Passing_reading_percentage_bySchool = round(Passing_reading_student_bySchool['reading_score'].count() / student_amount * 100,2)
Passing_reading_percentage_bySchool = Passing_reading_percentage_bySchool.to_frame(name='% Pass Reading')
per_school_summary = pd.merge(per_school_summary, Passing_reading_percentage_bySchool,on='school_name')
```

```python
Passing_maths_reading_student_df = Passing_reading_student_df.loc[school_data_complete_df['maths_score'] >= 50]
Passing_maths_reading_student_bySchool = Passing_maths_reading_student_df.groupby(['school_name'])
Passing_maths_reading_percentage_bySchool = round(Passing_maths_reading_student_bySchool['reading_score'].count() / student_amount * 100,2)
Passing_maths_reading_percentage_bySchool = Passing_maths_reading_percentage_bySchool.to_frame(name='% Overall Passing')
per_school_summary = pd.merge(per_school_summary, Passing_maths_reading_percentage_bySchool,on='school_name')
del per_school_summary['School ID']
per_school_summary.head(15)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
per_school_summary['category'] = pd.cut(per_school_summary['budget_average'], spending_bins, labels=labels, include_lowest=True)
school_spending_df = per_school_summary.sort_values(['category'])
```

```python
spending_maths_scores = school_spending_df.groupby(["category"]).mean()["average maths score"].to_frame(name='average maths score')
spending_reading_scores = school_spending_df.groupby(["category"]).mean()["average reading score"].to_frame(name='average reading score')
spending_summary_df = pd.merge(spending_maths_scores,spending_reading_scores , on='category' )
spending_passing_maths = school_spending_df.groupby(["category"]).mean()["% Pass Maths"].to_frame(name='% Pass Maths')
spending_summary_df = pd.merge(spending_summary_df,spending_passing_maths, on='category')
spending_passing_reading = school_spending_df.groupby(["category"]).mean()["% Pass Reading"].to_frame(name='% Pass Reading')
spending_summary_df = pd.merge(spending_summary_df,spending_passing_reading, on='category')
overall_passing_spending = school_spending_df.groupby(["category"]).mean()["% Overall Passing"].to_frame(name='% Overall Passing')
spending_summary_df = round (pd.merge(spending_summary_df,overall_passing_spending, on='category'),2)
spending_summary_df
```
